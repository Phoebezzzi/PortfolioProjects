```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import re
plt.rc('font',family = 'SimHei',size = 13)
import datetime
import seaborn as sns
```

```python
path = "boat_data.csv"
data = pd.read_csv(path)
data.head()
```

# data processing and cleaning

```python
data.shape
data.info()
```

```python
data.nunique()
```

```python
data['Currency'] = data.Price.str[:3]
data['Price'] = data.Price.str[4:].astype(int)
data = data.rename(columns = {'Number of views last 7 days':'Views'})
data.Currency.unique()
```

```python
## convert currency to USD
dict = {'CHF' : 1.01, 'EUR' : 1.06, 'DKK': 0.14, 'Â£ ' : 1.24}
data['Rate']=data.Currency
data.replace({'Rate': dict},inplace=True)
data['USD'] = data['Rate'] * data['Price']
data.head()
```

```python
## split the location into country, Region, Sub region
data[['Country', 'Region', 'Sub Region']] = data['Location'].str.split('» ',expand=True)
data['Country'] = data['Country'].str.replace(' Â', '')
data['Region'] = data['Region'].str.replace(' Â', '')
```

```python
## Handle TYpe Column
print(data.Type.unique())

def check_condition(x):
    if 'new' in str(x):
        y="New Boat"
    elif 'Used' in str(x):
        y = "Used Boat"
    else:
        y = "Unknown"
    return y

def fuel_type(x):
    if 'Diesel' in str(x):
        y="Diesel"
    elif ('Unleaded'or'Gas')in str(x):
        y = "Unleaded"
    elif 'Electric' in str(x):
        y = "Electric"
    elif 'Hybrid' in str(x):
        y = "Hybrid"
    else: 
        y = "Unknown"
    return y

data['Condition'] = data['Type'].apply(check_condition)
data['Fuel_Type'] = data['Type'].apply(fuel_type)
data['Fuel_Type'].value_counts()
data['Condition'].value_counts()
```

```python
# Handle Boat Type Column
data[['Boat Type-Main','Boat Type-2nd', 'Boat Type-3rd']] = data['Boat Type'].str.split(",",expand=True)
data.head()
```

```python
## handle Manufacturer
data["Manufacturer"] = data["Manufacturer"].replace({" power boats":""}, regex=True)
data.Manufacturer.value_counts()
```

```python
data_pop['Price_K']=data_pop.USD/1000.0
top200['Price_K']=top200.USD/1000.0
```

```python
data_pop.Price_K.describe()
```

```python
def PriceClassifiers(x):
    if x<50:
        z = '<50K'
        return z 
    elif x> 150:
        z = '>150K'
        return z 
    for y in range(50,150,50):
        if y<=x<y+50:
            z = str(y)+'-'+str(y+50)
            return z 
            break
        else:
            continue
data_pop['Price_Range'] = data_pop.Price_K.apply(PriceClassifiers)
top200['Price_Range'] = top200.Price_K.apply(PriceClassifiers)
```
